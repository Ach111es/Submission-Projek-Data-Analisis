## Visualization & Explanatory Analysis

### **Pertanyaan 1:** Bagaimana hubungan antara musim (season) dengan jumlah sewa sepeda (cnt)?

```python
plt.figure(figsize=(6,4), dpi=200)
ax = sns.barplot(day_df, x="season", y="cnt", estimator="sum", errorbar=None)
ax.set_ylim(0, 1200000)
ax.bar_label(ax.containers[0], fontsize=8)
```

**Wawasan:** Berdasarkan grafik di atas dapat dilihat bahwa penyewaan sepeda terbanyak terjadi pada musim ke-3 (musim gugur) dan penyewaan sepeda terendah terjadi pada musim ke-1 (musim semi)

### **Pertanyaan 2:** Bagaimana hubungan antara suhu (temp) dengan jumlah sewa sepeda (cnt)?

```python
plt.figure(figsize=(12,4), dpi=200)
graph = sns.scatterplot(x='temp', y='registered', data=day_df)
plt.title('Hubungan antara Suhu dan Pengguna Terdaftar')
plt.ylabel('Pengguna Terdaftar')
plt.xlabel('Suhu (temp)')
```

**Wawasan:** Berdasarkan grafik diatas terlihat terdapat korelasi positif antara suhu dengan jumlah pengguna yang terdaftar
